```python
import matplotlib.pyplot as plt
import matplotlib.image as pim
import numpy as np

%matplotlib notebook
```

```python
path = ''

```

```python
img = pim.imread(path + 'throne.png')

def show_img(title, _img):
    plt.figure(figsize=(8,6))
    plt.title(title)
    if _img.ndim == 2:
        plt.imshow(_img, cmap = 'gray')
    else:
        plt.imshow(_img)
    plt.show()

show_img("Throne", img)
```

```python
h, w = img.shape[:2]
print(img.shape, f'w={w}, h={h}')
```

```python
def lum(rgb):
    return np.dot(rgb, [0.2126, 0.7152, 0.0722])
#(0.299 * R, 0.587 * G, 0.114 * B)
#(0.2126 R + 0.7152 + G + 0.0722 B) 
gray = lum(img)
```

```python
print(gray.shape, img.shape)
```

```python
show_img("Throne", gray)
```

```python
x = np.arange(0, w, 1)
y = gray[300,:]

plt.figure(figsize=(8, 4))
plt.plot(x, y)
plt.show()
```

```python
from scipy import fftpack
im_fft = fftpack.fft(img)
im_fft
```

```python
print(img.shape, img.dtype)
print(im_fft.shape, im_fft.dtype)
```

```python
from matplotlib.colors import LogNorm

def plot_spectrum(im_fft, title):
    plt.title(title)
    plt.figure(figsize=(8,4))
    plt.imshow(np.abs(im_fft), norm=LogNorm(vmin=5))
    plt.colorbar()
    
plot_spectrum(im_fft, 'Fourier Transform')
```

```python
im_fft_shift = fftpack.fftshift(im_fft)
```

```python
plot_spectrum(im_fft_shift, 'Fourier Transform')
```
