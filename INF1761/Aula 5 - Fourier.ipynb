```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import fftpack
%matplotlib notebook
```

```python
path = '../dat/sgy/'
f3 = np.load(path + 'f3.npy')
_=plt.figure(figsize=[8,4])
plt.imshow(f3)
```

```python
f3.shape
```

```python
y = f3[:, 600]
dt = 4 / 1000
N = 363
T = N * dt
f = 1 / dt
t = np.arange(0, T, dt)
_= plt.figure(figsize=[8,4])
plt.plot(t,y)
plt.show()
print(f'sinal de {T} s com {N} amostras, dt={dt} e frequencia de amostragem = {f}')
```

```python
Y = fftpack.fft(y)

delta_f = 1 / T

fmax = N * delta_f

f = np.arange(0, fmax / 2, delta_f)

Y_shift = np.fft.fftshift(Y)

plt.figure(figsize=(8, 4))
plt.plot(f, np.abs(Y[: 182]))
plt.show()

```
