```python
import numpy as np

import pandas as pd

from ipycanvas import Canvas
```

```python
path = '../dat/cor/'
resp_path = 'resp/'
my_name = 'CG2020_T1_FelipeHolanda'
cie_table = pd.read_excel(path+'all_1nm_data.xls',header=3)
cie_table
```

```python
_lambda = cie_table['nm']
Lw = cie_table['CIE D65']
_x = cie_table['x bar']
_y = cie_table['y bar']
_z = cie_table['z bar']

_lambda = np.asarray(_lambda[80:481])
Lw = np.asarray(Lw[80:481])
_x = np.asarray(_x[80:481])
_y = np.asarray(_y[80:481])
_z = np.asarray(_z[80:481])

Lw = Lw / np.amax(Lw)

checker = pd.read_excel(path + 'ColorChecker_RGB_and_spectra.xls', 'spectral_data',skiprows=1,nrows=24)
checker
```

```python
reflectance = checker.loc[:, checker.columns[2:]].to_numpy(dtype=np.float32)
colors = checker['Color name']
```

```python
_Lw = Lw[(_lambda >379)&(_lambda<731)&(_lambda%10==0)]
xbar = _x[(_lambda>379)&(_lambda<731)&(_lambda%10==0)]
ybar = _y[(_lambda>379)&(_lambda<731)&(_lambda%10==0)]
zbar = _z[(_lambda>379)&(_lambda<731)&(_lambda%10==0)]
```

```python
k = 1.0 / (10 * np.sum(_Lw * ybar))

XYZ = np.zeros(shape = (colors.size, 3))

for i, cor in enumerate(colors):
    beta = reflectance[i, :]
    XYZ[i, 0] = 10 * k * np.sum(_Lw * beta * xbar)
    XYZ[i, 1] = 10 * k * np.sum(_Lw * beta * ybar)    
    XYZ[i, 2] = 10 * k * np.sum(_Lw * beta * zbar) 
      
```

```python
def gamma(val):
    if val < 0:
        return 0
    if (val < 0.0031308):      
        return 12.92 * val
    else:
        return 1.055 * val**(1.0 / 2.4) - 0.055

def CIEXYZ_to_sRGB(XYZ):
    Minv = np.array([[3.2404542,-1.5371385,-0.4985314],
                     [-0.9692660,1.8760108 ,0.0415560],
                     [ 0.0556434,-0.2040259,1.0572252]])
    rgb = np.dot(Minv,XYZ)
    sRGB = [gamma(i) for i in rgb]
    return sRGB
```

```python
def round_rect(canvas, x, y, w, h, r):
    canvas.begin_path()
    canvas.move_to(x + r, y)
    canvas.arc_to(x + w, y, x + w, y + h, r)
    canvas.arc_to(x + w, y + h, x, y + h, r)
    canvas.arc_to(x, y + h, x, y, r)
    canvas.arc_to(x, y, x + w, y, r)
    canvas.close_path()
```

```python
canvas = Canvas(width = 800, height = 600, sync_image_data = True)
```

```python
round_rect(canvas, 0, 0, 710, 510, 20)

canvas.fill_style = 'black'

canvas.fill()

idx = 0

sRGB_names = colors

_tuple = []

sRGB = list(_tuple)

for i in range(24):
    sRGB.append((int(255 * np.transpose(CIEXYZ_to_sRGB(XYZ[idx]))[0]), int(255 * np.transpose(CIEXYZ_to_sRGB(XYZ[idx]))[1]), int(255 * np.transpose(CIEXYZ_to_sRGB(XYZ[idx]))[2])))
    idx += 1
    
idx = 0

data = list([[]])

for i in range(4):
    for j in range(6):
        x = 30 + 110 * j
        y = 30 + 110 * i
        data.append(str(str("Color: " + sRGB_names[idx] + " - rgb" + str(sRGB[idx]) + "\n")))
        canvas.fill_style = 'rgb' + str(sRGB[idx])
        canvas.fill_rect(x, y, 100, 100)
        idx += 1 
            
        
np.save(resp_path + my_name + '.npy', data)
            
canvas.fill_style = 'white'

canvas.font = '30px serif'

canvas.fill_text('Felipe Holanda Bezerra', 210, 495)

canvas

canvas


```
